# Block 4: Wavelength, weather, and the scan
### Which wavelength, and which scan — and what does each choice cost?

Blocks 1–3 were about limits the radar can't escape: the beam spreads and climbs, one
measurement is a volume average, one pulse train can't beat cλ/8. This block is different —
it's about the **choices** a radar's designers and operators actually make, and the price the
physics charges for each. You pick a shorter wavelength for sensitivity and a finer beam from a
smaller dish; rain makes you pay in attenuation. You stack more tilts for vertical coverage; the
clock makes you pay in update time. Two halves — wavelength, then the scan — but one idea:
**every gain is bought with a coupled loss**, the same shape of trade you met in Block 1's
blockage and Block 3's dilemma. This block has three widgets. Run the setup cell once, then work
through them in order, since each builds on the last.

## Before we start
This is the synthesis block: every choice here is a trade, and every earlier idea resurfaces
inside one. Three things to hold onto.

1. **Wavelength is a choice with a price.** A shorter wavelength buys a finer beam from a
   smaller dish and more intrinsic sensitivity (Blocks 1–2) — but rain absorbs and scatters it
   more, so it attenuates. Negligible at S-band, real at C-band, severe at X-band.

2. **The two effects pull opposite ways.** A shorter wavelength is *both* more sensitive and
   more attenuated, and the attenuation penalty grows with range — so which band sees a faint
   distant storm is a competition, not a fixed answer.

3. **A scan is a time budget.**  Every beam position takes time to sample. Adding more tilts improves vertical coverage and structure, while spending longer on a tilt can improve data quality, but both slow the refresh rate of the volume. You can have more detail, cleaner measurements, or faster updates — but not all three at once.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Dropdown

C = 2.998e8; A_E = 6371.0
kB, T0 = 1.380649e-23, 290.0
ATTEN = {"S": (0.0001, 1.0), "C": (0.006, 1.0), "X": (0.03, 1.0)}  # k = a*R^b, dB/km one-way

def rain_scene(x_km, peak=80, bg=2, centre=40, width=8):
    return bg + peak*np.exp(-((x_km - centre)/width)**2)
def z_from_rain(R):
    return 10*np.log10(200 * np.maximum(R, 1e-6)**1.6)
def path_attenuation(x_km, R, band):
    a, b = ATTEN[band]; return 2*np.cumsum(a * R**b * np.gradient(x_km))
def beam_height(r_km, elev_deg):
    Re = (4/3)*A_E; phi = np.radians(elev_deg)
    return np.sqrt(r_km**2 + Re**2 + 2*r_km*Re*np.sin(phi)) - Re
def min_Z(r_km, power_kW=750, dish_m=8.5, freq_ghz=2.8, pulse_us=1.57,
          eta=0.45, NF_dB=4, Kw2=0.93, snr_dB=3):
    # minimum detectable reflectivity (dBZ) vs range; carries the 1/λ⁴ sensitivity at fixed dish
    lam = C/(freq_ghz*1e9); tau = pulse_us*1e-6; r = np.asarray(r_km, float)*1e3
    G = eta*(np.pi*dish_m/lam)**2; theta = 1.27*lam/dish_m
    N = kB*T0*(1/tau)*10**(NF_dB/10); snr = 10**(snr_dB/10)
    K = (np.pi**3 * C * power_kW*1e3 * G**2 * theta**2 * tau * Kw2)/(1024*np.log(2)*lam**2)
    return 10*np.log10((snr*N*r**2/K)/1e-18)

VCP = {"VCP 21 (9 tilts)":  [0.5,1.45,2.4,3.35,4.3,6.0,9.9,14.6,19.5],
       "VCP 12 (14 tilts)": [0.5,0.9,1.3,1.8,2.4,3.1,4.0,5.1,6.4,8.0,10.0,12.5,15.6,19.5],
       "VCP 31 clear-air (5)": [0.5,1.5,2.5,3.5,4.5],
       "fast low-level (3)": [0.5,1.5,2.5]}

print("Block 4 ready | attenuation across cell (dB):",
      {b: round(float(path_attenuation(np.linspace(0,120,600), rain_scene(np.linspace(0,120,600)), b)[-1]),1) for b in "SCX"},
      "| intrinsic MDS @100 km S/C/X:", [round(float(min_Z(100,freq_ghz=f)),1) for f in (2.8,5.6,9.4)])
```

## 1 — Shorter wavelength, bigger price: attenuation

A beam crosses a rain cell. Rain doesn't just reflect energy, it absorbs and scatters it *out* of the beam, so everything behind the cell is seen through a curtain of rain. That loss is negligible at S-band, real at C-band, and can wipe out the echo entirely at X-band.

```python
def plot_attenuation(peak_mmh=80):
    x = np.linspace(0, 120, 600); R = rain_scene(x, peak_mmh); z_true = z_from_rain(R)
    plt.figure(figsize=(8, 4.6))
    plt.plot(x, z_true, "k--", label="true reflectivity")
    for band, col in [("S", "tab:blue"), ("C", "tab:orange"), ("X", "tab:red")]:
        plt.plot(x, z_true - path_attenuation(x, R, band), color=col, lw=2, label=band + "-band observed")
    plt.ylim(-10, 60); plt.grid(alpha=0.3); plt.legend()
    plt.xlabel("range along beam (km)"); plt.ylabel("reflectivity (dBZ)")
    plt.title("attenuation carves a shadow behind the cell")
    plt.show()

interact(plot_attenuation,
         peak_mmh=FloatSlider(value=80, min=10, max=150, step=5, description="peak mm/h"));
```

**Basic**

1. Which band reads the cell peak closest to the truth? Which loses everything behind the core?
2. At 80 mm/h, roughly how many dBZ does the C-band echo lose behind the cell?

**A little further**

1. Compare the C-band profile with the S-band and X-band profiles as you increase the rain rate. Does C-band behave more like S-band or more like X-band? How does that change in heavy rain?
2. Raise the peak rain rate. Does the X-band shadow grow roughly in proportion to rain rate, or faster than proportionally? What does this imply about observing intense convective storms at X-band?

**Going deeper**

## 2 — Sensitivity vs attenuation: which band sees the distant storm?

Two wavelength effects pull in opposite directions. At the same dish, a *shorter* wavelength is intrinsically more sensitive — it can detect weaker echoes (the 1/λ⁴ in the radar equation, **dotted** lines). But a shorter wavelength also attenuates more, and that penalty *grows with range* (**solid** lines). Slide the rain the beam passes through and watch the curves cross: the band that sees a faint distant storm best is whichever wins this competition.

```python
def plot_detectability(path_rain_mmh=5):
    r = np.linspace(5, 250, 400)
    eff = {}
    plt.figure(figsize=(8, 4.8))
    for name, f, col in [("S", 2.8, "tab:blue"), ("C", 5.6, "tab:orange"), ("X", 9.4, "tab:red")]:
        a, b = ATTEN[name]
        intrinsic = min_Z(r, freq_ghz=f)                 # 1/λ⁴ sensitivity at the same dish
        eff[name] = intrinsic + 2*a*path_rain_mmh**b*r    # plus two-way path attenuation (dB)
        plt.plot(r, intrinsic, color=col, ls=":", lw=1, alpha=0.45)
        plt.plot(r, eff[name], color=col, lw=2.2, label=name + "-band")
    idx = np.where(np.diff(np.sign(eff["X"] - eff["S"])) > 0)[0]
    cross = r[idx[0]] if len(idx) else None
    if cross is not None:
        plt.axvline(cross, color="gray", ls="--", lw=0.8)
        ttl = "through " + str(path_rain_mmh) + " mm/h rain: X-band beats S-band only inside " + str(round(cross)) + " km"
    else:
        ttl = "no intervening rain: shorter wavelength wins at every range (1/λ⁴)"
    plt.ylim(-30, 50); plt.xlim(0, 250); plt.grid(alpha=0.3); plt.legend(loc="lower right")
    plt.xlabel("range (km)"); plt.ylabel("minimum detectable Z (dBZ)")
    plt.text(8, 44, "dotted: no-rain sensitivity · solid: through the rain", fontsize=8, color="gray")
    plt.title(ttl)
    plt.show()

interact(plot_detectability,
         path_rain_mmh=FloatSlider(value=5, min=0, max=20, step=1, description="path rain mm/h", style={'description_width': '150px'}, layout={'width': '500px'}));
```

**Basic**

2. Now raise the rain. Which band's curve climbs fastest, and which barely moves?

**A little further**

**Going deeper**

## 3 — Putting it together: scan strategies
A volume coverage pattern (VCP) is a stack of tilts built from a finite time budget. The fan (on Block 1's height axes) shows what each scanning choice buys and what it costs: adding tilts improves vertical coverage and shrinks gaps in the sampled atmosphere, but every additional tilt takes time, lengthening the interval between updates. The result is a three-way trade among the cone of silence above the highest tilt, the low-level gap below the lowest tilt, and the volume refresh time needed to complete the scan.

```python
def plot_vcp(pattern="VCP 21 (9 tilts)", rpm=3.0):
    tilts = VCP[pattern]; top, bottom = max(tilts), min(tilts); r = np.linspace(1, 300, 300)
    plt.figure(figsize=(8, 5.2))
    for e in tilts: plt.plot(r, beam_height(r, e), lw=1, color="tab:blue")
    plt.fill_between(r, beam_height(r, top), 20, color="tab:red", alpha=0.06)
    plt.text(35, 16.5, "cone of silence", color="tab:red", fontsize=9)
    plt.fill_between(r, 0, beam_height(r, bottom), color="gray", alpha=0.10)
    plt.text(150, 0.4, "low-level gap", color="dimgray", fontsize=9)
    vol = len(tilts)*60/rpm
    plt.xlim(0, 300); plt.ylim(0, 20); plt.grid(alpha=0.3)
    plt.xlabel("range (km)"); plt.ylabel("height (km)")
    plt.title(pattern + ": volume time " + str(round(vol)) + " s (" + str(round(vol/60,1))
              + " min)  |  cone half-angle " + str(round(90-top)) + "°")
    plt.show()

interact(plot_vcp,
         pattern=Dropdown(options=list(VCP), value="VCP 21 (9 tilts)", description="pattern"),
         rpm=FloatSlider(value=3, min=1, max=6, step=0.5, description="rpm"));
```

## 4 — Radar data: the wavelength trade, both ways

One figure, two pairs — each the same storm at the same time on an S-band and a C-band radar.
**Top row:** a heavy convective core (S-band left, C-band right). **Bottom row:** a lighter,
widespread echo (S-band left, C-band right; different case and a wider domain — note the change of scale). Together they
show the two sides of the wavelength trade from Widget 2.

For the bottom (sensitivity) pair, the C-band radar also runs a finer beam (0.5° vs 1°) and
more power (1 MW vs 500 kW) — so power, beamwidth, and wavelength all push sensitivity the same
way, and the pair on its own can't separate them. For the top (attenuation) pair, none of that
matters: attenuation is a loss the wave already suffered, and no amount of power or gain undoes
it — which is why C-band can lose behind the core even as the more capable radar.

**Basic**

1. Heavy-core pair: behind the strongest core, find where the C-band echo is weaker
   than the S-band. If both radars are calibrated and looking at the same storm, why do
   they report different reflectivities here? Note that C-band loses *even though* it's
   the more powerful radar with the finer beam.
2. Light-echo pair: which radar fills in the faint, widespread echo, and which breaks it into
   sparse speckle? Can we attribute this improvement just to the wavelength?

**A little further**

1. The detectability widget isolated wavelength by fixing the dish, and sensitivity went as
   1/λ⁴. These two radars differ in power, beamwidth, AND wavelength at once. With the
   beamwidth given, the radar-equation sensitivity is received power ∝ P_t / (θ²·λ²) — so more
   power, a finer beam, and a shorter wavelength all push it the same way. Can this pair, on
   its own, isolate how much of C-band's advantage is the wavelength?
2. Estimate the C-band deficit behind the core (compare the dBZ category on each at the same
   spot). Explain whether the C-band radar's extra power or finer beam reduce that deficit. What would the loss do to a rainfall estimate downrange?

**Going deeper**

1. The two pairs aren't equally clean. Attenuation reads straight off the heavy-core images:
   it's a path loss on the wave, and the radar constant (P_t, gain, λ) is calibrated out of a
   detected echo's dBZ — extra power or gain can't undo a loss the signal already suffered.
   The sensitivity "win" can't be read that cleanly, because wavelength helps sensitivity too
   (the 1/λ⁴ from the widget) and is tangled with power and beam. Explain why the widget
   (fixed dish) can separate them and these two radars cannot.
2. Quantify it from P_r ∝ P_t/(θ²·λ²): power gives C-band +3 dB (2×), the finer beam +6 dB
   (gain ∝ 1/θ², so 0.5° vs 1° is 4×), and the shorter wavelength another +6 dB (1/λ², with
   λ_S/λ_C ≈ 2) — about +15 dB. Now reconcile with the widget's 1/λ⁴: at fixed
   dish θ ∝ λ/D, so 1/(θ²λ²) becomes D²/λ⁴ — the λ⁴ already contains the finer beam. Notice the
   two beamwidths are in the same 2:1 ratio as the wavelengths, so the dishes are about equal
   (~8 m): this is the fixed-dish case, the finer beam is the wavelength's doing, and the only
   separate advantage is the 2× power. Given everything stacks for C-band, why does the
   heavy-core attenuation conclusion still stand?

*In practice, dual-polarization C-band radars recover much of this lost reflectivity using
specific differential phase (KDP), a phase measurement that attenuation doesn't affect. That topic is the subject of Block 5 (dual-polarization), and it's why the shadow you see here isn't the end of the
story for operational C-band networks.*

## Block 4 takeaway
**Every choice carries a coupled cost. A shorter wavelength buys a finer beam and more sensitivity from a smaller dish but pays in attenuation that grows with range, so which band sees a distant storm is a competition between the two; and stacking tilts into a volume buys vertical coverage but trades it against clean velocity and update speed.**
